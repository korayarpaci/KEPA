### 1 günlük altın

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

# Örnek tarih ve altın fiyatı verileri
tarihler = ['2024-12-21 09:00', '2024-12-21 12:00', '2024-12-21 15:00', '2024-12-21 18:00', '2024-12-21 21:00']
fiyatlar = [1850.5, 1851.0, 1850.8, 1852.2, 1851.9]

# Tarihleri datetime formatına dönüştür
tarihler = [datetime.strptime(tarih, '%Y-%m-%d %H:%M') for tarih in tarihler]

# Grafik oluşturma
plt.figure(figsize=(10, 5))
plt.plot(tarihler, fiyatlar, marker='o', linestyle='-', color='gold')  # Altın için 'gold' rengi kullanıldı
plt.gcf().autofmt_xdate()
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m %H:%M'))
plt.title('Altın Fiyatları - Son 1 Gün')
plt.xlabel('Tarih')
plt.ylabel('Fiyat (USD/ons)')
plt.grid(True)
plt.show()
```

## 1 haftalık gold

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import random

# Simulate 1-week gold price data (mock data for demonstration)
timestamps = [datetime.now() - timedelta(hours=6 * i) for i in range(28)]  # 7 days, 4 data points per day
prices = [random.uniform(1900, 2000) for _ in range(28)]

# Create a DataFrame for the mock data
gold_data_week = pd.DataFrame({
    'Timestamp': timestamps,
    'Price': prices
})

# Sort the DataFrame by timestamp for accurate plotting
gold_data_week = gold_data_week.sort_values(by='Timestamp')

# Plotting the 1-week gold price chart
plt.figure(figsize=(14, 7))
plt.plot(gold_data_week['Timestamp'], gold_data_week['Price'], label='Gold Price', marker='o')
plt.title('Gold Price - 1 Week', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Price (USD)', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()

# Optionally, display the DataFrame
print(gold_data_week)
```

### 1 yıllık gold

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import random

# Simulate 1-year gold price data (mock data for demonstration)
timestamps = [datetime.now() - timedelta(weeks=i) for i in range(52)]  # 52 weeks, 1 data point per week
prices = [random.uniform(1800, 2100) for _ in range(52)]

# Create a DataFrame for the mock data
gold_data_year = pd.DataFrame({
    'Timestamp': timestamps,
    'Price': prices
})

# Sort the DataFrame by timestamp for accurate plotting
gold_data_year = gold_data_year.sort_values(by='Timestamp')

# Plotting the 1-year gold price chart
plt.figure(figsize=(14, 7))
plt.plot(gold_data_year['Timestamp'], gold_data_year['Price'], label='Gold Price', marker='o')
plt.title('Gold Price - 1 Year', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Price (USD)', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()

# Optionally, display the DataFrame
print(gold_data_year)
```

```python
import json
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, timedelta
import os
import random

# Örnek altın fiyatı verileri (1 yıllık)
tarihler = [(datetime.now() - timedelta(days=i)).strftime('%Y-%m-%d') for i in range(364, -1, -1)]  # 1 yıl (365 gün)
fiyatlar = [random.uniform(1800, 2100) for _ in range(365)]  # 365 adet rastgele fiyat (1800-2100 USD)

# Tarihleri datetime formatına dönüştür
tarihler_datetime = [datetime.strptime(tarih, '%Y-%m-%d') for tarih in tarihler]

# Grafiği çiz
plt.figure(figsize=(10, 5))
plt.plot(tarihler_datetime, fiyatlar, marker='o', linestyle='-', color='gold')  # Altın için 'gold' rengi kullanıldı
plt.gcf().autofmt_xdate()
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
plt.title('Altın Fiyatları - Son 1 Yıl')
plt.xlabel('Tarih')
plt.ylabel('Fiyat (USD/ons)')
plt.grid(True)
plt.show()

# JSON formatında veriyi hazırlama
veri = {
    "gold_fiyatlari": [
        {"tarih": tarih, "fiyat": round(fiyat, 2)}  # Fiyatları yuvarlayarak kaydet
        for tarih, fiyat in zip(tarihler, fiyatlar)
    ]
}

# Dosya yolunu belirleme
klasor_yolu = "/Users/koray/Desktop/grafikler"
dosya_yolu = os.path.join(klasor_yolu, "1yeargold.json")

# Klasör yoksa oluştur
os.makedirs(klasor_yolu, exist_ok=True)

# JSON dosyasını kaydet
with open(dosya_yolu, "w", encoding="utf-8") as dosya:
    json.dump(veri, dosya, ensure_ascii=False, indent=4)

print(f"1 yıllık altın fiyatları başarıyla {dosya_yolu} yoluna kaydedildi.")
```

### 5 yıllık gold

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import random
import os

# Simulate 5-year gold price data (mock data for demonstration)
timestamps = [datetime.now() - timedelta(days=30 * i) for i in range(60)]  # 60 months (5 years), 1 data point per month
prices = [random.uniform(1700, 2200) for _ in range(60)]

# Create a DataFrame for the mock data
gold_data_5years = pd.DataFrame({
    'Timestamp': timestamps,
    'Price': prices
})

# Sort the DataFrame by timestamp for accurate saving and plotting
gold_data_5years = gold_data_5years.sort_values(by='Timestamp')

# Convert Timestamp to string for JSON serialization
gold_data_5years['Timestamp'] = gold_data_5years['Timestamp'].astype(str)

# Define the save path
save_path = '/Users/koray/Desktop/grafikler/gold'

# Ensure the directory exists
os.makedirs(save_path, exist_ok=True)

# Save the DataFrame as a JSON file
file_name = '5yeargold.json'
full_path = os.path.join(save_path, file_name)
gold_data_5years.to_json(full_path, orient='records', lines=True)

# Confirmation message
print(f"File saved to: {full_path}")
```
