```python
import json
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, timedelta
import os
import random

# Örnek gümüş fiyatı verileri (1 günlük)
tarihler = [datetime.now().strftime('%Y-%m-%d')]
fiyatlar = [random.uniform(22, 30)]  # 1 adet rastgele fiyat (22-30 USD)

# Grafiği çiz
plt.figure(figsize=(10, 5))
plt.plot(tarihler, fiyatlar, marker='o', linestyle='-', color='silver')  # Gümüş için 'silver' rengi kullanıldı
plt.gcf().autofmt_xdate()
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
plt.title('Gümüş Fiyatları - 1 Günlük')
plt.xlabel('Tarih')
plt.ylabel('Fiyat (USD/ons)')
plt.grid(True)
plt.show()

# JSON formatında veriyi hazırlama
veri = {
    "silver_fiyatlari": [
        {"tarih": tarih, "fiyat": round(fiyat, 2)}  # Fiyatları yuvarlayarak kaydet
        for tarih, fiyat in zip(tarihler, fiyatlar)
    ]
}

# Dosya yolunu belirleme
klasor_yolu = "/Users/koray/Desktop/grafikler/silver"
dosya_yolu = os.path.join(klasor_yolu, "1daysilver.json")

# Klasör yoksa oluştur
os.makedirs(klasor_yolu, exist_ok=True)

# JSON dosyasını kaydet
with open(dosya_yolu, "w", encoding="utf-8") as dosya:
    json.dump(veri, dosya, ensure_ascii=False, indent=4)

print(f"1 günlük gümüş fiyatları başarıyla {dosya_yolu} yoluna kaydedildi.")
```
